1. tentukan library yang digunakan

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

2. load dataset

```python
survey_lung_cancer_dataset = pd.read_csv('survey lung cancer.csv')
```

```python
survey_lung_cancer_dataset.head()
```

```python
survey_lung_cancer_dataset['GENDER']=survey_lung_cancer_dataset['GENDER'].map({'M':1, 'F':0})
survey_lung_cancer_dataset
```

```python
survey_lung_cancer_dataset['LUNG_CANCER']=survey_lung_cancer_dataset['LUNG_CANCER'].map({'YES':1, 'NO':0})
survey_lung_cancer_dataset
```

```python
survey_lung_cancer_dataset.shape
```

```python
survey_lung_cancer_dataset['LUNG_CANCER'].value_counts()
```

```python
# memisahkan data dan label
X = survey_lung_cancer_dataset.drop(columns='LUNG_CANCER', axis=1)
Y = survey_lung_cancer_dataset['LUNG_CANCER']
```

```python
print (X)
```

```python
print (Y)
```

3. standarisasi data

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data=scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X=standarized_data
Y=survey_lung_cancer_dataset['LUNG_CANCER']
```

```python
print (X)
print(Y)
```

4. memisahkan data training dan data testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

5. membuat data latih menggunakan algoritma SVM

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

6. membuat model evaluasi untuk mengukur tingkat akurasi

```python
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print('Akurasi data training adalah=', training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print('Akurasi data testing adalah=', test_data_accuracy)
```

7. membuat model prediksi

```python
input_data=(0, 59, 1, 1, 1, 2, 1, 2, 1, 2, 1, 2, 2, 1, 2)

input_data_as_numpy_array=np.array(input_data)

input_data_reshape=input_data_as_numpy_array.reshape(1,-1)

std_data=scaler.transform(input_data_reshape)
print(std_data)

prediction=classifier.predict(std_data)
print(prediction)

if(prediction[0]==0):
    print('Pasien tidak terkena Lung Cancer')
else :
    print('Pasien terkena lung cancer')
```
